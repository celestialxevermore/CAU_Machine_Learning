```python
import numpy as np
import matplotlib.image as img
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.colors as colors
```

```python
I0 = img.imread('test.jpeg')
```

```python
# ++++++++++++++++++++++++++++++++++++++++++++++++++
# complete the blanks
#
num_row     = I0.shape[0]
num_column  = I0.shape[1]
num_channel = I0.shape[2]
#
# ++++++++++++++++++++++++++++++++++++++++++++++++++

print('number of rows of I0 = ', num_row)
print('number of columns of I0 = ', num_column)
print('number of channels of I0 = ', num_channel)
```

```python
# ++++++++++++++++++++++++++++++++++++++++++++++++++
# complete the blanks
#
I = (I0[:,:,0] + I0[:,:,1] + I0[:,:,2])/3
    
    
num_row     = I.shape[0]
num_column  = I.shape[1]
#
# ++++++++++++++++++++++++++++++++++++++++++++++++++

print('number of rows of I = ', num_row)
print('number of columns of I = ', num_column)
```

```python
# ++++++++++++++++++++++++++++++++++++++++++++++++++
# complete the blanks
#

I = (I - I.min())/(I.max()-I.min())

#
# ++++++++++++++++++++++++++++++++++++++++++++++++++

print('maximum value of I = ', np.max(I))
print('minimum value of I = ', np.min(I))
```

```python
def compute_derivative_x_forward(I):
    
    D = np.zeros(I.shape)
    
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    
    I1 = I[1:,:]
    I2 = np.vstack((I1,I[-1,:]))
    
    D = I2 - I
    
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    
    return D
```

```python
def compute_derivative_x_backward(I):
    
    D = np.zeros(I.shape)
    
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    
    I1 = I[0,:]
    I2 = np.vstack((I1,I[0:-1,:]))
    
    D = I - I2
    
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    
    return D
```

```python
def compute_derivative_x_central(I):
    
    D = np.zeros(I.shape)
    
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    
    a1 = I[1:,:]
    a2 = np.vstack((a1,I[-1,:]))
    
    b1 = I[0,:]
    b2 = np.vstack((b1,I[0:-1,:]))
    
    D = (a2 - b2)/2
    
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    
    return D
```

```python
def compute_derivative_y_forward(I):
    
    D = np.zeros(I.shape)
    
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    I1 = I[:,1:]
    I2 = np.hstack((I1,I[:,-1:]))
    
    D = I2 - I
    
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    
    return D
```

```python
def compute_derivative_y_backward(I):
    
    D = np.zeros(I.shape)
    
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #

    I1 = I[:,0:1]
    I2 = np.hstack((I1,I[:,0:-1]))
    
    D = I - I2
    
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    
    return D
```

```python
def compute_derivative_y_central(I):
    
    D = np.zeros(I.shape)
    
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    
    a1 = I[:,1:]
    a2 = np.hstack((a1,I[:,-1:]))
    
    b1 = I[:,0:1]
    b2 = np.hstack((b1,I[:,0:-1]))

    D = (a2 - b2)/2
    
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    
    return D
```

```python
def compute_norm_gradient_central(I):
    
    norm_gradient = np.zeros(I.shape)
    
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    a1 = I[1:,:]
    a2 = np.vstack((a1,I[-1,:]))
    
    b1 = I[0,:]
    b2 = np.vstack((b1,I[0:-1,:]))
    
    c = (a2 - b2)/2
    
    d1 = I[:,1:]
    d2 = np.hstack((d1,I[:,-1:]))
    
    e1 = I[:,0:1]
    e2 = np.hstack((e1,I[:,0:-1]))

    f = (d2 - e2)/2
    
    norm_gradient = c*c + f*f
    
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    
    return norm_gradient
```
