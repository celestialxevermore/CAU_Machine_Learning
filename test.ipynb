```python
def compute_derivative_x_forward(I):

    dd = np.zeros(I.shape)
    D = np.zeros(I.shape)

    for i in range(1,len(I)):
        dd[i,:] = I[i-1,:]
    dd[0,:] = I[0,:]
    print(dd)
    D = I-dd 
    return D
```

```python
def compute_derivative_x_backward(I):
    D = np.zeros(I.shape)
    temp = np.zeros(I.shape)

    for i in range(len(I)-1):
        temp[i,:] = I[i+1,:]
    temp[-1,:] = I[-1,:]

    D = I-temp 
    return D
```

```python
res = compute_derivative_x_backward(aa)
print(res)
```

```python
def compute_derivative_x_central(I):

    D = np.zeros(I.shape)

    tmp = np.zeros([I.shape[0]+2,I.shape[1]])
    tmp[0] = I[0,:]
    tmp[-1] = I[-1,:]

    for i in range(len(I)):
        tmp[i+1] = I[i,:]
    
    for i in range(1,len(tmp)-1):
        D[i-1] = tmp[i+1,:]-tmp[i-1,:]
    return D
```

```python
res = compute_derivative_x_central(aa)
print(res)
```
